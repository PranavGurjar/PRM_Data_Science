```python
# Ordinal Encoding -> 1. Cyclic Line , 2. Map Function 
```

```python
# Using Cyclic Line
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size" : ["s", "m", "l", "xl", "s", "m", "l", "s", "s", "l", "xl", "m"]})
df.head(3)
```

```python
ord_data1 = [["s", "m", "l", "xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data1)
oe.fit(df[["Size"]])
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
```

```python
df
```

```python
# Using Map Function
```

```python
ord_data2 = {"s":0, "m":1, "l":2, "xl":3}
```

```python
df["Size_en_map"] = df["Size"].map(ord_data2)
```

```python
df
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset["Property_Area"].unique()
```

```python
dataset["Property_Area"].fillna(dataset["Property_Area"].mode()[0], inplace=True)
```

```python
en_data_ord = [['Rural', 'Semiurban', 'Urban']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oen = OrdinalEncoder(categories=en_data_ord)
```

```python
dataset["Property_Area"] = oen.fit_transform(dataset[["Property_Area"]])
```
