```python
# Kernal Function
# 1. Linear: K(w, b) = ((w^T) * x) + b
# 2. Polynomial: K(w, x) = ((γ * (w^T) * x) + b^N)
# 3. Gaussian RBF: K(w, x) = exp(-γ || xi - xj || ^n)
# 4. Sigmoid: K(xi, xj) = tanh(α * xi^T * xj + b)
```

```python
dataset = pd.read_csv("nonlinear_data_circles.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="data1", y="data2", data=dataset, hue="output")
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset.iloc[:, :-1]
x
```

```python
# y = dataset["placement"]
# y
```

```python
y = dataset["output"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel="linear")
sv.fit(x_train, y_train)
```

```python
# Kernel type:
#               'linear', 'poly', 'rbf', 'sigmoid'
```

```python
sv.score(x_train, y_train)*100
```

```python
sv.score(x_test, y_test)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=sv)
plt.show()
```
