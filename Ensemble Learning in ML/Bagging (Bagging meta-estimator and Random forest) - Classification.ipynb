```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x, y = make_moons(n_samples=1000, noise=0.2)
```

```python
x
```

```python
y
```

```python
df = {"x1":x[:,0], "x2":x[:,1],"y":y}
```

```python
dataset = pd.DataFrame(df)
```

```python
dataset
```

```python
sns.scatterplot(x="x1", y="x2", data=dataset, hue="y")
plt.show()
```

```python
x_a = dataset.iloc[:, :-1]
y_a = dataset["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_a, y_a, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg = BaggingClassifier(estimator=SVC(), n_estimators=50)
bg.fit(x_train, y_train)
```

```python
bg.score(x_train, y_train)*100, bg.score(x_test, y_test)*100
```

```python
svc = SVC()
svc.fit(x_train, y_train)
```

```python
svc.score(x_train, y_train)*100, svc.score(x_test, y_test)*100
```

```python
rfc = RandomForestClassifier(n_estimators=30)
rfc.fit(x_train, y_train)
```

```python
rfc.score(x_train, y_train)*100, rfc.score(x_test, y_test)*100
```
