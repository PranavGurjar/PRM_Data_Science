```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
dataset = pd.read_csv("placements.csv")
dataset.head(3)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["package"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_train, y_train)*100, lr.score(x_test, y_test)*100
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
dt.score(x_train, y_train)*100, dt.score(x_test, y_test)*100
```

```python
svr = SVR()
svr.fit(x_train, y_train)
svr.score(x_train, y_train)*100, svr.score(x_test, y_test)*100
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("lr1", LinearRegression()), ("dt1", DecisionTreeRegressor()), ("svr1", SVR())]
```

```python
vr = VotingRegressor(li, weights=[10, 20, 30])
vr.fit(x_train, y_train)
```

```python
vr.score(x_train, y_train)*100, vr.score(x_test, y_test)*100
```

```python
df = {"lr":lr.predict(x_test), "dt":dt.predict(x_test), "svr":svr.predict(x_test), "vr":vr.predict(x_test)}
```
