```python
# Imbalanced Dataset Handling
```

```python
# 1. random under sampling
# 2. random over sampling
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
dataset.head(3)
```

```python
dataset["Purchased"].value_counts()
```

```python
x = dataset.iloc[:,:-1]
x
```

```python
y = dataset["Purchased"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[19, 19000]])
```

```python
lr.predict([[45, 26000]])
```

# imblearn

# RandomUnderSampler

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru = RandomUnderSampler()
ru_x, ru_y = ru.fit_resample(x, y)
```

```python
ru_x
```

```python
ru_y
```

```python
ru_x.value_counts()
```

```python
ru_y.value_counts()
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(ru_x, ru_y, test_size=0.2, random_state=42)
```

```python
lr1 = LogisticRegression()
lr1.fit(x_train1, y_train1)
```

```python
lr1.score(x_test1, y_test1)*100
```

```python
lr1.predict([[19, 19000]])
```

```python
lr1.predict([[45, 26000]])
```

```python
lr1.predict([[46, 28000]])
```

# RandomOverSampler

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro = RandomOverSampler()
ro_x, ro_y = ro.fit_resample(x, y)
```

```python
ro_x
```

```python
ro_y
```

```python
ro_y.value_counts()
```

```python
x_train2, x_test2, y_train2, y_test2 = train_test_split(ro_x, ro_y, test_size=0.2, random_state=42)
```

```python
lr2 = LogisticRegression()
lr2.fit(x_train2, y_train2)
```

```python
lr2.score(x_test2, y_test2)*100
```

```python
lr2.predict([[19, 19000]])
```

```python
lr2.predict([[45, 26000]])
```

```python
lr2.predict([[46, 28000]])
```
