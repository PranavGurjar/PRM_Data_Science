```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Iris.csv")
dataset.head(3)
```

```python
dataset["Species"].unique()
```

```python
sns.pairplot(data=dataset, hue="Species")
plt.show()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
x = dataset.iloc[:, :-1]
x
```

```python
y = dataset["Species"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# ovr

```python
from sklearn.linear_model import LogisticRegression
```

```python
# lr = LogisticRegression(multi_class="ovr")
# lr.fit(x_train, y_train)
```

```python
# lr.score(x_test, y_test)
```

```python
from sklearn.multiclass import OneVsRestClassifier
```

```python
lr_ovr = OneVsRestClassifier(LogisticRegression())
lr_ovr.fit(x_train, y_train)
```

```python
lr_ovr.score(x_test, y_test)*100
```

# multinomial

```python
# lr1 = LogisticRegression(multi_class="multinomial")
# lr1.fit(x_train, y_train)
```

```python
# lr1.score(x_test, y_test)
```

```python
# You no longer need the multi_class argument. 
# The model detects multiple classes and uses 'multinomial' by default.
lr1 = LogisticRegression(solver="lbfgs", max_iter=1000)
lr1.fit(x_train, y_train)
```

```python
lr1.score(x_test, y_test)*100
```

# LogisticRegression

```python
lr2 = LogisticRegression()
lr2.fit(x_train, y_train)
```

```python
lr2.score(x_test, y_test)*100
```
