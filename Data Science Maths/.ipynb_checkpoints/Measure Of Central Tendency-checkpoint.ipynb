```python
# using the titanic.csv dataset for performing mathematical means operations
```

```python
dataset = pd.read_csv("titanic.csv")
```

```python
dataset.head(3)
```

```python
dataset["Age"].mean()
```

```python
mn = np.mean(dataset["Age"])
```

```python
sns.histplot(x="Age", data=dataset, bins=[i for i in range(0, 81, 10)])
plt.plot([mn for i in range(0, 300)], [i for i in range(0, 300)], c = "red")       # mn = np.mean(dataset["Age"])
plt.show()
```

```python
# 2. Median
```

```python
dataset.isnull().sum()
```

```python
dataset["Age"].fillna(dataset["Age"].mean(), inplace = True)
```

```python
np.median(dataset["Age"])
```

```python
dataset["Age"].median()
```

```python
mdf = np.median(dataset["Fare"])
```

```python
dataset["Fare"].median()
```

```python
dataset.isnull().sum()
```

```python
mnf = np.mean(dataset["Fare"])
```

```python
sns.histplot(x = "Fare", data = dataset, bins = [i for i in range(0, 81, 10)])
plt.plot([mnf for i in range(0, 300)], [i for i in range(0, 300)], c = "red")
plt.plot([mdf for i in range(0, 300)], [i for i in range(0, 300)], c = "blue")
plt.show()
```

```python
# 3. Mode
```

```python
dataset["Fare"].mode()
```

```python
type(dataset["Fare"].mode())
```

```python
mof = dataset["Fare"].mode()[0]
```

```python
type(dataset["Fare"].mode()[0])
```

```python
dataset["Fare"].value_counts()
```

```python
sns.histplot(x = "Fare", data = dataset, bins = [i for i in range(0, 81, 10)])
plt.plot([mnf for i in range(0, 300)], [i for i in range(0, 300)], c = "green", label = "mean")
plt.plot([mdf for i in range(0, 300)], [i for i in range(0, 300)], c = "blue", label = "median")
plt.plot([mof for i in range(0, 300)], [i for i in range(0, 300)], c = "red", label = "mode")
plt.legend()
plt.show()
```
