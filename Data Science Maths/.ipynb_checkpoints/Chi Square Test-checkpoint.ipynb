```python
# Hypothesis Testing
# 3. Chi Square Test
```

```python
# Example 1
```

```python
import numpy as np
from scipy.stats import chi2
```

```python
ob = np.array([22, 17, 20, 26, 22, 13])
ex = np.array([20, 20, 20, 20, 20, 20])
```

```python
chi_cal = np.sum(np.square(ob - ex)/ex)
chi_cal
```

```python
# significance level (alpha)
alpha = 0.05  

# degrees of freedom
df = 5  

# chi-squared critical value
chi_table = chi2.ppf(1 - alpha, df)
```

```python
chi_table
```

```python
# Example 2
```

```python
row1 = np.array([40, 45, 25, 10])
row2 = np.array([35, 30, 20, 30])
```

```python
row1, row2
```

```python
s_no = np.sum(row1) + np.sum(row2)
s_no
```

```python
sum_r1 = np.sum(row1)
sum_r2 = np.sum(row2)
```

```python
s_no, sum_r1, sum_r2
```

```python
sum_row = np.array([sum_r1, sum_r2])
sum_col = row1 + row2
sum_row, sum_col
```

```python
for i in sum_row : 
    for j in sum_col : 
        print(i, j)
```

```python
exp = []
for i in sum_row : 
    for j in sum_col : 
        value = (i * j)/s_no
        exp.append(value)
```

```python
exp
```

```python
obj = np.array([40, 45, 25, 10, 35, 30, 20, 30])
obj
```

```python
len(obj), len(exp)
```

```python
chi_cal = np.sum(np.square(obj - exp)/exp)
chi_cal
```

```python
# significance level (alpha)
alpha = 0.05  

# degrees of freedom
# df = (row - 1) * (col - 1)
df = (2 - 1) * (4 - 1)

# chi-squared critical value
chi_table = chi2.ppf(1 - alpha, df)
```

```python
alpha, df, chi_table
```
