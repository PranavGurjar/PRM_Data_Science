```python
# 2. Mean Absolute Deviation
```

```python
# Mean
```

```python
sec_a = np.array([75, 65, 73, 68, 72, 76])
sec_b = np.array([90, 47, 43, 96, 93, 51])
no = np.array([1, 2, 3, 4, 5, 6])
```

```python
sec_a, sec_b, no
```

```python
meansec_a = np.mean(sec_a)
meansec_b = np.mean(sec_b)
meanseci_b = int(np.mean(sec_b))
```

```python
meansec_a, meansec_b, meanseci_b
```

```python
mean_b = int(sec_b.mean())
```

```python
mean = np.mean(sec_b)
```

```python
mean
```

```python
plt.figure(figsize = (12, 6))
plt.scatter(sec_a, no, c = "blue", label = "sec A")
plt.scatter(sec_b, no, c = "red", label = "sec B")
plt.plot([70, 70, 70, 70, 70, 70], no, label = "Mean")
plt.legend()
plt.show()
```

```python
# Mean Absolute Deviation. 
```

```python
np.abs(-2)
```

```python
sec_a - mean
```

```python
sec_b - mean
```

```python
np.abs(sec_a - mean)
```

```python
np.abs(sec_b - mean)
```

```python
np.sum(np.abs(sec_a - mean))
```

```python
np.sum(np.abs(sec_b - mean))
```

```python
mad_a = np.sum(np.abs(sec_a - mean))/len(sec_a)
mad_b = np.sum(np.abs(sec_b - mean))/len(sec_b)
```

```python
mad_a, mad_b
```
