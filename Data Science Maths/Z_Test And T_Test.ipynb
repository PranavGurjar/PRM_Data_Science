```python
# Example 1
```

```python
import numpy as np
import scipy.stats as st
```

```python
s_x = 90
p_u = 82
p_std = 20
n = 81
ap = 0.05
```

```python
z_cal = (s_x - p_u) / (p_std / np.sqrt(n))
z_cal
```

```python
z_table = st.norm.ppf(1 - ap)
z_table
```

```python
if z_table < z_cal :
    print("ha is right")
else : 
    print("ho is right")
```

```python
# Before: 50, 60, 45, 65, 55, 70, 40, 75, 80, 65, 70, 60, 50, 55, 45, 75, 60, 50, 65, 70
# After: 60, 70, 55, 75, 65, 80, 50, 85, 90, 70, 75, 65, 55, 60, 50, 80, 65, 55, 70, 75 
```

```python
t_table = st.t.ppf(1 - 0.025, 19)
t_table
```

```python
Before = np.array([50, 60, 45, 65, 55, 70, 40, 75, 80, 65, 70, 60, 50, 55, 45, 75, 60, 50, 65, 70])
After = np.array([60, 70, 55, 75, 65, 80, 50, 85, 90, 70, 75, 65, 55, 60, 50, 80, 65, 55, 70, 75])
Before, After
```

```python
mean_a = np.mean(After)
mean_b = np.mean(Before)
std_a = np.std(After)
std_b = np.std(Before)
no_a = len(After)
no_b = len(Before)

mean_a, mean_b, std_a, std_b, no_a, no_b
```

```python
t_cal = (mean_a - mean_b) / (np.sqrt(((std_a * std_a)/no_a) + ((std_b * std_b)/no_b)))
t_cal
```

```python
if t_cal > t_table :
    print("ha is right")
else :
    print("ho is right")
```
