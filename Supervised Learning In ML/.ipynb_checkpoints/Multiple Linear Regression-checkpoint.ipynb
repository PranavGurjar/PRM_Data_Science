```python
# Multiple Linear Regression
```

```python
# Multiple Linear Regression is an extension of Simple Linear regression as it takes more than one predictor variable to predict the response variable.
```

```python
# y = m1x1 + m2x2 + m3x3 + c 
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("employee.csv")
dataset.head(3)
```

```python
dataset['Salary'] = dataset['Salary'].str.replace(',', '').astype(float)
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
sns.heatmap(data=dataset.corr(), annot=True)
plt.show()
```

```python
x = dataset.iloc[:,:-1]
x
```

```python
y = dataset["Salary"]
y
```

```python
x.ndim
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```
