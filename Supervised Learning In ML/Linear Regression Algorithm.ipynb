```python
# Linear equation : y = mx + c
# where:
# y = dependent variable
# x = independent varible
# m = slope/gradient/coeffient
# c = intercept
```

```python
# m = +ve ---- @ < 90
# m = -ve ---- @ > 90
# m = 0 ---- @ = 0
```

```python
# c = +ve
# c = -ve
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("placement.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(5, 4))
sns.scatterplot(x = "cgpa", y = "package", data = dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
# x = dataset[["cgpa"]]
# x.ndim
```

```python
# y = dataset[["package"]]
# y.ndim
```

```python
x = dataset[["cgpa"]]
y = dataset["package"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
# 6.89	3.26
```

```python
lr.predict([[6.89]])
```

```python
# y = mx + c
# y = lr.coef_ * x + lr.intercept_
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# y = 0.57425647*x-1.0270069374542108 
```

```python
0.57425647*6.89-1.0270069374542108 
```

```python
y_prd = lr.predict(x)
```

```python
plt.figure(figsize=(5, 4))
sns.scatterplot(x = "cgpa", y = "package", data = dataset, label = "org data")
plt.plot(dataset["cgpa"], y_prd, c = "red", label = "predict line")
plt.legend(["org data", "predict line"])
plt.savefig("predict.jpg")
plt.show()
```
